# stemming

## stemming is the process of reducing a word to its root word

example: actor, actresse, acting = act

````python
port_stem = PorterStemmer()
````

````python
def stemming (content):
  stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  return ' '.join(stemmed_content)
````

````python
df['stemmed_text'] = df['text'].apply(stemming)
````

````python
# Locate NaN in DataFrame
nan_indices = df[df['stemmed_text'].isna()].index
print(nan_indices)
````

````python
df['target'].value_counts()
````

**now** split and Vectorize: Split your data again and apply the vectorizer:

````python
# lets update our variables
X = df['stemmed_text'].values
Y = df['target'].values

# splitting the dataset to training & test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
````

````python
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
````

````python
print('x_train= ',X_train)
print('x_test= ', X_test)
````

Training the machine learning model

# logistic regression

````python
model = LogisticRegression(max_iter=1000)
````

````python
model.fit(X_train,  Y_train)
````

Model Evaluation

accuracy score

````python
# accuracy score on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
````

````python
print('Accuracy score on the training data: ',training_data_accuracy)
````

````python
# accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
````

````python
print('Accuracy score on the test data: ',test_data_accuracy)
````

we can see test and train accuracy are very near to each Other
# model accuracy = 77.87%
